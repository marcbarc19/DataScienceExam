```python
import pandas as pd
from functools import reduce
```

```python
df_list = [population_df, consize_df, VA_df, PS_df, GE_df, RQ_df, RL_df, CC_df, EDU_df, INV_df, SI_df, ICT_df]

# Convert 'Year' column to numeric for each DataFrame in df_list
for df in df_list:
    if 'Year' in df.columns:
        df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Identify the key columns for merging
key_columns = ['Country Name', 'Country Code', 'Year']  # Example key columns

# Merge the DataFrames
predictor_df = df_list[0]  # Initialize merged DataFrame with the first DataFrame
for df in df_list[1:]:
    predictor_df = pd.merge(predictor_df, df, on=key_columns, how='outer', suffixes=('_left', '_right'))

# Replace '..' with NaN
predictor_df.replace('..', pd.NA, inplace=True)

# Fill NaN values with 'NaN' if needed
predictor_df.fillna('NaN', inplace=True)

# Drop rows where 'Year' column is 2023
predictor_df = predictor_df[predictor_df['Year'] != 2023]

display(predictor_df)
```
